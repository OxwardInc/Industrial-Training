```python
file = open('text_file.txt', 'r')

content = file.read()

value = content.split()

longest_word=''


for word in value:
    if len(word) > len(longest_word):
        longest_word = word
    
length_of_longest_word = len(longest_word)
print(f"The longest word's are {longest_word} and its lenght is {length_of_longest_word}")

```

```python
import random

print("SN \t Total \t Simulated percentage \t Expected percentage")
print('*' * 100)

for i in range(1000):
    sn = i + 1
    x = random.randint(1, 7)
    y = random.randint(1, 7)
    
    value = x + y

    
    print(f"{sn} \t {value}")
```

```python
empty_list = []

while True:
    number_entered = int(input("enter the numbers:"))
    if number_entered == 0:
        break
    empty_list.append(number_entered)
    
    empty_list.sort()
    
print(empty_list)

for value in empty_list:
    print(f"{value}")
```

```python
empty_list = []

while True:
    number_entered = int(input("enter the numbers:"))
    if number_entered == 0:
        break
    empty_list.append(number_entered)
    
    empty_list.sort(reverse=True)
    
print(empty_list)

for value in empty_list:
    print(f"{value}")
```

```python
stored_data = []
while True:
    value_entered = input("enter anything:")
    if value_entered == '':
        break
    else:
        if value_entered in stored_data:
            pass
        else:
            stored_data.append(value_entered)
    
print(stored_data)

for value in stored_data:
    print(f"{value}")
```
