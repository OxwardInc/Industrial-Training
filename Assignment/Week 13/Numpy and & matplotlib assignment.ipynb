### Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ages = np.random.randint(18, 50, 9) #Generating 9 values between 18 & 50
```

```python
ages
```

```python
names = ['Maria', 'Charles', 'Ezekiel', 'Malv', 'Whyte', 'Charity', 'Oxward', 'Vicent', 'Glory']
plt.hist(ages)
```

```python
plt.hist(ages);
```

```python
plt.hist(ages, bins= 10);
```

```python
plt.scatter(ages, names);
```

```python
plt.boxplot(ages, names);
```
