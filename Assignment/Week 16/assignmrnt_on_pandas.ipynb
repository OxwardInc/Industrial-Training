#### Largest consumption of alcohol or drink

```python
# Find the maximum value for each column
max_beer_servings = data_frame['beer_servings'].max()
max_spirit_servings = data_frame['spirit_servings'].max()
max_wine_servings = data_frame['wine_servings'].max()
max_total_litres_of_pure_alcohol = data_frame['total_litres_of_pure_alcohol'].max()

# Filter the DataFrame to get the rows with the maximum values
max_beer_country = data_frame[data_frame['beer_servings'] == max_beer_servings]
max_spirit_country = data_frame[data_frame['spirit_servings'] == max_spirit_servings]
max_wine_country = data_frame[data_frame['wine_servings'] == max_wine_servings]
max_total_litres_of_pure_alcohol_country = data_frame[data_frame['total_litres_of_pure_alcohol'] == max_total_litres_of_pure_alcohol]

# Combine the filtered DataFrames
new_data = pd.concat([max_beer_country, max_spirit_country, max_wine_country, max_total_litres_of_pure_alcohol_country])

# Replace non-maximum values with NaN
new_data.loc[new_data['beer_servings'] != max_beer_servings, 'beer_servings'] = np.nan
new_data.loc[new_data['spirit_servings'] != max_spirit_servings, 'spirit_servings'] = np.nan
new_data.loc[new_data['wine_servings'] != max_wine_servings, 'wine_servings'] = np.nan
new_data.loc[new_data['total_litres_of_pure_alcohol'] != max_total_litres_of_pure_alcohol, 'total_litres_of_pure_alcohol'] = np.nan

new_data.head()
```

##### Least consumption of alcohol or drink

```python
min_beer_servings = data_frame['beer_servings'].min()
min_spirit_servings = data_frame['spirit_servings'].min()
min_wine_servings = data_frame['wine_servings'].min()
min_total_litres_of_pure_alcohol_servings = data_frame['total_litres_of_pure_alcohol'].min()

min_beer_country = data_frame[data_frame['beer_servings'] == min_beer_servings]
min_spirit_country = data_frame[data_frame['spirit_servings'] == min_spirit_servings]
min_wine_country = data_frame[data_frame['wine_servings'] == min_wine_servings]
min_total_litres_of_pure_alcohol_country  = data_frame[data_frame['total_litres_of_pure_alcohol'] == min_total_litres_of_pure_alcohol_servings]

new_data = pd.concat([min_beer_country, min_spirit_country, min_wine_country, min_total_litres_of_pure_alcohol_country])

new_data.loc[new_data['beer_servings'] != min_beer_servings, 'beer_servings'] = np.nan
new_data.loc[new_data['spirit_servings'] != min_spirit_servings, 'spirit_servings'] = np.nan
new_data.loc[new_data['wine_servings'] != min_wine_servings, 'wine_servings'] = np.nan
new_data.loc[new_data['total_litres_of_pure_alcohol'] != min_total_litres_of_pure_alcohol_servings, 'total_litres_of_pure_alcohol'] = np.nan

new_data.head()
```
