## Filling by interpolation

### Linear Interpolation:

##### Linear interpolation estimates missing values by drawing a straight line between two known data points and calculating the value of the missing point along that line. It assumes a constant rate of change between adjacent data points.

```python
import pandas as pd
```

```python
# Creating a data frame with missing values
df = pd.DataFrame({'A': [1, 2, None, 4, None, 6]})
```

```python
# filling by liner interpolation

df_linear_interpolation_data = df.interpolate(method='linear')
```

```python
df_linear_interpolation_data
```

### Quadratic Interpolation:

##### Quadratic interpolation fits a quadratic curve between two adjacent data points and estimates the missing value based on the curve. It assumes a quadratic relationship between data points.

```python
df_quad_interpolate = df.interpolate(method='quadratic')
```

```python
df_quad_interpolate
```

### Cubic Interpolation:

##### Cubic interpolation fits a cubic spline between two adjacent data points and estimates the missing value based on the spline. It assumes a smooth curve between data points.

```python
df_cubid_interp = df.interpolate(method='cubic')
```

```python
df_cubid_interp
```

### Time-based Interpolation:

##### If your DataFrame has a time-based index, you can use time-based interpolation methods like time or pad to fill missing values based on time intervals.

```python
df_time_interp = df.interpolate(method='pad')
```

```python
df_time_interp
```
