```python
from tkinter import *

root = Tk()

frame1 = Frame(root)
frame1.pack()

frame2 = Frame(root)
frame2.pack()

Label(frame1, text = "Number one").pack(side = 'left')

Entry1 = Entry(frame1)
Entry1.pack()

Label(frame2, text = "Number two").pack(side = 'left')

Entry2 = Entry(frame2)
Entry2.pack()

Button(text = 'Add', command = compute).pack()

root.mainloop()
```

```python

def add_numbers():
    print(f"Entry 1: {Entry1.get()}")
    print(f"Entry 2: {Entry2.get()}")
    compute = float(Entry1.get()) + float(Entry2.get())
    print(f"Solution: {compute}")

def sub_numbers():
    print(f"Entry 1: {Entry1.get()}")
    print(f"Entry 2: {Entry2.get()}")
    compute = float(Entry1.get()) - float(Entry2.get())
    print(f"Solution: {compute}")

def mul_numbers():
    print(f"Entry 1: {Entry1.get()}")
    print(f"Entry 2: {Entry2.get()}")
    compute = float(Entry1.get()) * float(Entry2.get())
    print(f"Solution: {compute}")

def div_numbers():
    print(f"Entry 1: {Entry1.get()}")
    print(f"Entry 2: {Entry2.get()}")
    compute = float(Entry1.get()) / float(Entry2.get())
    print(f"Solution: {compute}")
```

```python
from tkinter import *

root = Tk()

frame1 = Frame(root)
frame1.pack()

frame2 = Frame(root)
frame2.pack()

Label(frame1, text = "Number one").pack(side = 'left')

Entry1 = Entry(frame1)
Entry1.pack()

Label(frame2, text = "Number two").pack(side = 'left')

Entry2 = Entry(frame2)
Entry2.pack()

Button(text = 'Add', command = add_numbers).pack(side = 'left')
Button(text = 'Sub', command = sub_numbers).pack(side = 'right')
Button(text = 'Mul', command = mul_numbers).pack(side = 'left')
Button(text = 'Div', command = div_numbers).pack(side = 'right')

root.mainloop()
```
