```python
plt.pie(ages, labels= names, autopct= '%.2f%%', explode=(0.2, 0, 0, 0, 0, 0, 0, 0, 0));
```

```python
plt.pie(ages, labels= names, autopct= '%.2f%%', explode=(0.2, 0, 0, 0.6, 0, 0, 0, 0, 0));
```

```python
plt.pie(ages, labels= names, autopct= '%.2f%%', explode=(0.2, 0, 0, 0, 0.5, 0, 0.3, 0, 0));
```

```python
plt.pie(ages, labels= names, autopct= lambda value: value / sum(ages), explode=(0.2, 0, 0, 0, 0.5, 0, 0.3, 0, 0));
```

```python
plt.pie(ages, labels= names, autopct= lambda value: round(value / sum(ages), 1), 
explode=(0.2, 0, 0, 0, 0.5, 0, 0.3, 0, 0));
```

```python
plt.pie(ages, labels= names, autopct= lambda value: round(360 * value / sum(ages), 1), 
explode=(0.2, 0, 0, 0, 0.5, 0, 0.3, 0, 0));
```

```python
plt.pie(ages, labels= names, autopct= lambda value: round(360 * value / 100, 1), 
explode=(0.2, 0, 0, 0, 0.5, 0, 0.3, 0, 0));
```

```python
plt.pie(ages, labels= names, autopct= lambda value: f"{round(360 * value / 100, 1)}$^0$", 
explode=(0.2, 0, 0, 0, 0.5, 0, 0.3, 0, 0));
```
